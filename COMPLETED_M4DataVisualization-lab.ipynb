## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT *
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

df['ConvertedComp'].plot(kind='hist')
```

Plot a box plot of `Age.`


```python
df['Age'].plot(kind='box')
```

## Visualizing relationships in data

Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
df.plot.scatter(x='Age', y='WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
plt.scatter(x='WorkWeekHrs', y='CodeRevHrs', s='Age', alpha=0.5, data=df)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY2 = """
SELECT *
FROM DatabaseDesireNextYear
"""

df_db = pd.read_sql_query(QUERY2,conn)

df_grouped = df_db.groupby('DatabaseDesireNextYear', axis=0).count()
df_grouped.sort_values(['Respondent'], ascending=False, inplace=True)
df_top5 = df_grouped.head()
print(df_top5)

df_top5['Respondent'].plot(kind='pie', autopct='%.1f')

# Set title and remove 'Respondents' label on y-axis
plt.title('Top 5 Databases that Respondents Wish to Learn Next Year')
ax = plt.gca()
ax.get_yaxis().set_visible(False)
plt.show()
```

```python
# FROM QUIZ: In the list of most popular languages respondents wish to learn
# next year, what is the rank of Python?

QUERY4 = """
SELECT *
FROM LanguageDesireNextYear
"""

df_lang = pd.read_sql_query(QUERY4, conn)

df_lang2 = df_lang.groupby('LanguageDesireNextYear').count()
df_lang2.sort_values(['Respondent'], ascending=False, inplace=True)
df_lang2.head()
```

```python
# FROM QUIZ: How many respondents indicated that they currently work with 'SQL'?

QUERY5 = """
SELECT *
FROM LanguageWorkedWith
"""

df_lang_sql = pd.read_sql_query(QUERY5, conn)

df_lang_sql = df_lang_sql.groupby('LanguageWorkedWith').count()
df_lang_sql.loc['SQL']
```

```python
# FROM QUIZ: How many respondents indicated that they work on 'MySQL' only?

QUERY6 = """
SELECT Respondent, DatabaseWorkedWith, Count(*) AS Count
FROM DatabaseWorkedWith
GROUP BY Respondent
HAVING Count = 1 
   AND DatabaseWorkedWith = 'MySQL'
"""
df_mysql = pd.read_sql_query(QUERY6 ,conn)
print('Number of respondents who said they only work with MySQL:',
      df_mysql.shape[0])
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY3 = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df_sc = pd.read_sql_query(QUERY3, conn)

df_sc_gb = df_sc.groupby('Age').median().sort_values('Age')
print(df_sc_gb)
df_sc_gb.plot.bar(stacked=True, ylabel='Hours Workded',
                  title='Median Work Week Hours and Code Revision Hours by Age')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY_LC = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df_lc = pd.read_sql_query(QUERY_LC, conn)

df_lc = df_lc.groupby('Age').median()

df_lc.plot(ylabel='Hours Workded',
           title='Median Converted Compensation for Ages 45 to 60')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY_BC = """
SELECT MainBranch
FROM master
"""

df_bc = pd.read_sql_query(QUERY_BC, conn)
df_bc_cnt = df_bc.value_counts()
print(df_bc_cnt)
df_bc_cnt.plot(kind='barh')

plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch Location')
plt.title('Developer vs Non-Developer')
plt.show
```

```python
# FROM QUIZ: Majority of the survey responders are:
# Developer - mobile; Developer - front-end; Developer - full-stack; Developer - back-end
# df_lang2.sort_values(['Respondent'], ascending=False, inplace=True)

QUERY_DEVTYPE = """
SELECT *
FROM DevType
"""

df_devtype = pd.read_sql_query(QUERY_DEVTYPE, conn)

df_devtype_gb = df_devtype.groupby('DevType').count()
df_devtype_gb.sort_values(['Respondent'], ascending=True, inplace=True)
df_devtype_gb.plot(kind='barh')

plt.title('Developer Type of Respondents')
plt.xlabel('Number of Respondents')
plt.ylabel('Developer Type')
plt.show()
```

```python
# FROM QUIZ: Based on the Line Chart of Age and median ConvertedComp for the
# age group 25-30, what kind of correlation do you see between Age and
# median ConvertedComp?

QUERY_LC2 = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 25 AND 30
"""

df_lc2 = pd.read_sql_query(QUERY_LC2, conn)

df_lc2 = df_lc2.groupby('Age').median()

df_lc2.plot(ylabel='Hours Workded',
           title='Median Converted Compensation for Ages 25 to 30')

df_lc2.corr()
```
